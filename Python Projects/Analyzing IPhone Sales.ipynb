# <center><span>SALES DATA ANALYSIS</span></center>

```python
# HIGHEST RATED PRODUCTS

highest_rated = file.sort_values(by='Star Rating',ascending=False)[:10]
highest_rated
```

```python
highest_rated_product = highest_rated['Product Name']
highest_rated_product
```

```python
iphone_names = highest_rated['Product Name'].value_counts()
iphone_names
```

```python
label = iphone_names.index
label
```

```python
counts = highest_rated['Number Of Ratings']
counts.index
```

```python
plt.barh(label,counts)
plt.show()
```

```python
# PRODUCTS WITH THE HIGHEST REVIEWS
reviews = highest_rated['Number Of Reviews']

plt.barh(label, reviews)
plt.show()
```

```python
# RELATIONSHIP BETWEEN SALES PRICE AND NUMBER OF REVIEWS

plt.scatter(reviews, label, marker='x')
plt.show()
```

```python
# RELATIONSHIP BETWEEN SALES PRICE AND STAR RATINGS

plt.scatter(counts, label, marker='x')
plt.show()
```
