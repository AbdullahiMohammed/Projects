## Analysis of Used Car listings
This project aims to clean and analyze a  used car dataset obtained from eBay Kleinanzeigen, a classifieds section of the German eBay website.

Looking at the data, we have some cleaning tasks to do. Columns that dont have much uniqueness about them have to be dropped, columns that are in text format instead of numeric need to be converted.

```python
autos.drop(['seller', 'offer_type', 'nr_of_pictures'], axis=1)
```

```python
autos['price']  = autos['price'].str.replace('$', '').str.replace(',', '').astype(int)
```

```python
autos['odometer'] = autos['odometer'].str.replace(',', '').str.replace('km', '').astype(int)
```

```python
autos.rename(columns = {'odometer':'odometer_km'}, inplace=True)
```

### The maximum year is 9999 and the minimum year is 1000, this isnt acceptable so we have to further clean the year column to a  more realistic timeline it 1900 - 2016, 2016 being the most current year in the dataset

```python
autos = autos[autos['registration_year'].between(1900, 2016)]
```

```python
autos['registration_year'].describe()
```

```python
autos['brand'].unique()
```

```python
#autos['brand'].value_counts(normalize=True)
top_brand = autos['brand'].value_counts(normalize=True)
top_brands = top_brand[top_brand > 0.05].index
top_brands
```

```python
top_brands_mean_price = {}

for brand in top_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_price = brand_only['price'].mean()
    top_brands_mean_price[brand] = int(mean_price)
    
top_brands_mean_price
```

```python
mileage = {}

for brand in top_brands:
    brand_only = autos[autos['brand'] == brand]
    miles_mean = brand_only['odometer_km'].mean()
    mileage[brand] = int(miles_mean)
    
mileage
```

```python
m_prices = pd.Series(top_brands_mean_price).sort_values(ascending=False)
m_prices
```

```python
m_mileage = pd.Series(mileage).sort_values(ascending=False)
m_mileage
```

```python
d_price = pd.DataFrame(m_prices, columns=['Average Price'])
d_price
```

```python
d_mileage = pd.DataFrame(m_mileage, columns=['Average Mileage'])
d_mileage
```
