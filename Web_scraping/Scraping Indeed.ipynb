```python
import pandas as pd #to handle data and its analysis
import requests #to get webpages
from bs4 import BeautifulSoup  #to parse webpages
import datetime #to format time
```

```python
def extract(page):
    
    url = 'https://www.glassdoor.com/Job/nigeria-data-analysts-jobs-SRCH_IL.0,7_IN177_KO8,21.htm?clickSource=searchBox'
    r = requests.get(url).text
    soup = BeautifulSoup(r, 'html.parser')
    return soup
joblist = []

def transform(soup):
    cards = soup.find_all('li',  'eigr9kq0')
    for item in cards:
        title = item.find('a', 'eigr9kq1').text
        company = item.find('a', 'e1n63ojh0').text
        jobage = item.find(attrs = {'data-test': 'job-age'}).text
        location = item.find('span','e1rrn5ka4').text
        job_description = 'https://www.glassdoor.com' + soup.find('a', 'jobLink').get('href')
        
        job = {
            'Title' : title,
            'Company': company,
            'Job Age' : jobage,
            'Location' : location,
            'Job Description Link': job_description
        }
        joblist.append(job)
    return


for i in range(0,40,10):
    print('Getting Page ' + str(i))
    d = extract(3)
    transform(d)

df = pd.DataFrame(joblist)
df.to_csv('jobslist.csv')
```
